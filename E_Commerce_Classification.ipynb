```python
ls = (df.isnull().mean() * 100)[(df.isnull().mean() * 100)>0].keys().tolist()
cols_num = [c for c in ls if df[c].dtype in ['int64', 'float64']]

cols_obj = [c for c in ls if df[c].dtype == 'object']

print("Colonnes numriques dtectes :", cols_num)
print("Colonnes 'object' dtectes :", cols_obj)
```

```python
df[cols_num].head()
```

```python
cols_object = df.select_dtypes(include='object').keys().tolist()
```

```python
df_final = pd.get_dummies(df, columns=cols_object, drop_first=True)

df_final.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = df_final.corr()
fig, ax = plt.subplots(figsize=(18, 12))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    center=0,
    ax=ax
)

plt.tight_layout()
```

```python
corr_target = df_final.corr()['Churn'].abs()
cols_to_keep = corr_target[corr_target > 0.05].index.tolist()
df_final = df_final[cols_to_keep]
```
